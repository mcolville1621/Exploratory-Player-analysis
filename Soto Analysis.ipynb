```python
HOF_cols = HOF[["Name","BA", "OBP", "SLG", "OPS", "K_rate", "BB_rate", "SB_rate"]]
HOF_cols
```

```python
bat = batting_stats(2018,2022, qual = 40, ind=0)
Soto = bat.query('Name == "Juan Soto"')
Soto_stats = Soto[["Name","AVG", "OBP", "SLG", "OPS", "SO", "BB", "PA", "SB", "CS"]]
Soto_stats["K_rate"] = Soto_stats["SO"]/ Soto_stats["PA"]
Soto_stats["BB_rate"] = Soto_stats["BB"]/ Soto_stats["PA"]
Soto_stats["SB_rate"] = Soto_stats["SB"]/ (Soto_stats["SB"] + Soto_stats["CS"])
```

```python
Soto_stats["BA"] = Soto_stats["AVG"]
Soto_stats.drop(columns = ["PA", "AVG", "BB", "SB", "SO", "CS"], inplace = True)
Soto_stats
```

```python
Combined = pd.concat([Soto_stats, HOF_cols])
Combined = Combined.query('Name not in @ls')
Combined.sort_values("OBP", ascending = False).head(50)
```
